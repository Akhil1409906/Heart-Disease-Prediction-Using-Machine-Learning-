```python
x = df.drop('target', axis=1)  # Features
y = df['target']  # Target variable
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x, y, test_size = 0.2, random_state=50)
```

```python
from sklearn.ensemble import AdaBoostClassifier
# Initialize the AdaBoostClassifier
model = AdaBoostClassifier()

# Fit the model with training data
model.fit(x_train, y_train)

# Print the model details
print(model)
```

```python
y_pred=model.predict(x_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score,classification_report
print("TRAIN accuracy score:",model.score(x_train,y_train))
```

```python
print("TEST accuracy score:",accuracy_score(y_pred,y_test))
```

```python
import pickle
with open('xgboost.pkl', 'wb') as file:
    pickle.dump(model, file)
```

```python
def predict_with_manual_input():
    # Collect user input
    print("Enter values for the following features:")
    age = float(input("Age: "))
    sex = int(input("Sex (0 or 1): "))
    cp = int(input("CP (chest pain type): "))
    trestbps = float(input("Resting blood pressure: "))
    chol = float(input("Serum cholesterol: "))
    fbs = int(input("Fasting blood sugar (0 or 1): "))
    restecg = int(input("Resting electrocardiographic results: "))
    thalach = float(input("Maximum heart rate achieved: "))
    exang = int(input("Exercise induced angina (0 or 1): "))
    oldpeak = float(input("Oldpeak: "))
    slope = int(input("Slope of peak exercise ST segment: "))
    ca = int(input("Number of major vessels colored by fluoroscopy: "))
    thal = int(input("Thalassemia: "))

    # Create a DataFrame for the input
    input_data = pd.DataFrame({
        'age': [age],
        'sex': [sex],
        'cp': [cp],
        'trestbps': [trestbps],
        'chol': [chol],
        'fbs': [fbs],
        'restecg': [restecg],
        'thalach': [thalach],
        'exang': [exang],
        'oldpeak': [oldpeak],
        'slope': [slope],
        'ca': [ca],
        'thal': [thal]
    })

    # Make prediction
    prediction = model.predict(input_data)
    print(f'Predicted target value: {prediction[0]}')

# Call the function to make a prediction
predict_with_manual_input()
```
